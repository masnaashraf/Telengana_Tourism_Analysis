# Provide Insights For Telangana Government Tourism Department

The Telangana Government Tourism Department is responsible for promoting and developing tourism in the Indian state of Telangana. Its main objectives are to attract tourists to Telangana, create tourism infrastructure, and develop new tourism products. The department offers a variety of tourism services and facilities such as accommodation, transportation, and tour packages. It also organizes various events and festivals throughout the year to showcase the state's culture and heritage.

Meta Data

This file contains all the meta information regarding the columns described in the CSV files.

1.domestic_visitors.csv: contains data about visitors who travel within the country(India) to 'Telangana' districts for various purposes, such as tourism, business, visiting friends and family.

2.foreign_visitors.csv: contains data about international visitors, referring to individuals who travel from one country to another for various purposes such as tourism, business, education, or visiting friends and family.

Column Description for domestic_visitors:

district: The name of the district in Telangana.
date: The starting date of the month when the tourist data was collected for the specified district.
month: The month for which the tourist data is being reported.
year: The year is in YYYY format.
visitors: The number of domestic visitors who visited the given district in Telangana during the specified month and year.
      This could include tourists, business travellers, and other types of domestic visitors.
Column Description for foreign_visitors:

district: The name of the district in Telangana.
date: The starting date of the month when the tourist data was collected for the specified district.
month: The month for which the tourist data is being reported.
year: The year is in YYYY format.
visitors: The number of foreign visitors who visited the given district in Telangana during the specified month and year.

# 2.	List down the top 3 districts based on compounded annual growth rate (CAGR) of visitors between (2016 - 2019)?

**(Insight: Districts that are growing)**


To calculate the compounded annual growth rate (CAGR) of visitors for each district, we can use the following formula:

**CAGR = ((EV / BV) ^ (1/n)) - 1**

**where EV = Ending Value, BV = Beginning Value, and n = number of years.**

We can calculate the beginning value as the number of visitors in 2016, the ending value as the number of visitors in 2019, and n as 4 (since we are calculating the CAGR over a period of 4 years).

Here's the code to calculate the CAGR for each district:

**calculating for domestic visitors**

```python
# Group data by district and sum the visitors for each year
visitors_by_year_and_domesticdistrict = domestic_visitors.groupby(['year', 'district'])['visitors'].sum().reset_index()
visitors_by_year_and_domesticdistrict
```

```python
# Pivot the data to have year as columns and district as rows
visitors_pivot_domestic = visitors_by_year_and_domesticdistrict.pivot(index='district', columns='year',values='visitors')
visitors_pivot_domestic
```

```python
# Calculate the CAGR for each district
visitors_pivot_domestic['CAGR'] = (((visitors_pivot_domestic[2019]/visitors_pivot_domestic[2016]) ** (1/4)) - 1) * 100
```

```python
visitors_pivot_domestic
```

```python
# Sort the data in descending order and select top 3 districts
top_3_districts_domesticvisitors= visitors_pivot_domestic.sort_values('CAGR', ascending=False).head(3)
top_3_districts_domesticvisitors.to_clipboard()
top_3_districts_domesticvisitors
```

**calculating for foriegn visitors**

```python

foreign_visitors.head()
```

```python
# Group data by district and sum the visitors for each year
visitors_foriegn_by_year_and_district = foreign_visitors.groupby(['year', 'district'])['visitors'].sum().reset_index()
visitors_foriegn_by_year_and_district
```

```python
# Pivot the data to have year as columns and district as rows
visitorsforiegn_pivot = visitors_foriegn_by_year_and_district.pivot(index='district', columns='year',values='visitors')
visitorsforiegn_pivot
```

```python
# Calculate the CAGR for each district
visitorsforiegn_pivot['CAGR'] = (((visitorsforiegn_pivot[2019]/visitorsforiegn_pivot[2016]) ** (1/4)) - 1) * 100
```

```python
import pandas as pd

# Define the available data
years = [2016, 2017, 2018, 2019]
domestic_visitors = [23394705, 27160242, 19543651, 13802362]

# Calculate the compounded annual growth rate (CAGR)
start_value = domestic_visitors[0]
end_value = domestic_visitors[-1]
num_years = len(years)
CAGR = (end_value / start_value) ** (1/num_years) - 1

# Use the CAGR to project the number of domestic tourists in Hyderabad for each year from 2016 to 2025
start_year = 2016
end_year = 2025
projected_visitors = []
for year in range(start_year, end_year+1):
    if year in years:
        projected_visitors.append(domestic_visitors[years.index(year)])
    else:
        num_years = year - start_year
        projected_visitors.append(start_value * (1 + CAGR) ** num_years)

# Create a DataFrame to store the projected visitors for each year
projected_domestic = pd.DataFrame({'Year': range(start_year, end_year+1),
                             'Projected Visitors': projected_visitors})

# Print the projected visitors DataFrame
print(projected_domestic)

projected_domestic.to_clipboard()
```
